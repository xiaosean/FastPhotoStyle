```python
from __future__ import print_function
import argparse
import torch
import process_stylization
from photo_wct import PhotoWCT
from photo_smooth import Propagator
```

```python
from os import path
from os import listdir
```

```python
model ='./PhotoWCTModels/photo_wct.pth'
save_intermediate = False
fast = True
no_post = False
cuda = 0
```

```python
p_pro = Propagator()
```

# setup model

```python
p_wct = PhotoWCT()
```

```python
p_wct.load_state_dict(torch.load(model))
```

```python
p_wct.cuda(0)
```

```python
content_image_path ='./demo_image_fastphoto/gta/00016.png'
content_seg_path =[]
style_image_path ='./demo_image_fastphoto/city/hamburg_000000_042885_leftImg8bit.png'
style_seg_path = []
conten_filename = content_image_path.split("/")[-1][:-4]
style_filename = style_image_path.split("/")[-1][:-4]
out_filename = []
output_image_path = './demo_image_fastphoto/result/%s_%s.png' % (conten_filename, style_filename)
# output_image_path = path.join(output_image_root, item['name'])
```

```python
print("output_image_path =", output_image_path)

process_stylization.stylization(
    stylization_module=p_wct,
    smoothing_module=p_pro,
    content_image_path=content_image_path,
    style_image_path=style_image_path,
    content_seg_path=[],
    style_seg_path=[],
    output_image_path=output_image_path,
    cuda=1,
    save_intermediate=False,
    no_post=True
)
```

```python
content_dir_path = "./demo_image_fastphoto/gta"
style_dir_path ='./demo_image_fastphoto/city'
save_dir_path ='./demo_image_fastphoto/result'
```

```python
for content_im in listdir(content_dir_path):
    if ".png" in content_im :
        print(content_im)
        for style_im in listdir(style_dir_path):
            if ".png" in style_im :
                print(style_im)
                output_im = content_im[:-4] + "_" + style_im
                process_stylization.stylization(
                    stylization_module=p_wct,
                    smoothing_module=p_pro,
                    content_image_path=path.join(content_dir_path, content_im),
                    style_image_path=path.join(style_dir_path, style_im),
                    content_seg_path=[],
                    style_seg_path=[],
                    output_image_path=path.join(save_dir_path, output_im),
                    cuda=1,
                    save_intermediate=False,
                    no_post=True
                )
```
